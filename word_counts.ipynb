```python
import pandas as pd
```

```python
df_private = pd.read_csv("private_db.csv")
df_public = pd.read_csv("public_database.csv")
df_public.head()
```

```python
'''
Creates data frame with unique words from given keywords
'''
# df to hold word and number of occurances
keywords = pd.DataFrame(columns = ['keyword', 'occurances'])
word_index = 1 #index to track progress of adding words

# check if each word is already in df, add if not present
for entry in df_public.head(100)['keywords']:
    word_bag = str(entry).split('; ')
    for word in word_bag:
        if not(word.lower in list(keywords['keyword'])):
            keywords.loc[-1] = [word.lower(), 0] #add word with 0 initial occurances
            keywords.index = keywords.index + 1
            if word_index % 1000 == 0: # display progress every 1000
                print("Adding word " + str(word_index))
            word_index += 1
keywords
```

```python
#string cleanup into list
keyword_string = '"peptide", "heart", "pah", "laminae"'
keyword_string = keyword_string.replace('\"', '')
keyword_string = keyword_string.replace(',', '')
keyword_list = keyword_string.split()
keyword_list

#make df and set keyword
keywords = pd.DataFrame({
    "keyword": keywords_list,
    "occurances": [0] * len(keywords_list)
})

keywords = keywords.set_index("keyword")
```

```python
keyword_num = 1
#loop through each keyword in dataframe
for keyword in list(keywords.index):
    word_count = 0
    print("Working on word " +  str(keyword_num))
    
    #loop through each company description
    for description in list(df_public['abstract_text']):
        words = str(description).split() #split string into list of words
        
        #count number of keyword occurances in entry
        for word in words:
            
            if word.lower() == keyword.lower():
                word_count += 1
    keywords.loc[keyword, 'occurances'] = word_count
    keyword_num +=1

#print df sorted by number of occurances
print(keywords.sort_values(by='occurances', ascending=False))
```
